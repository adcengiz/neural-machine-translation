```python
plt.figure(figsize=(20,10))
plt.plot([[*bleu_no_beam][k] for k in [int(x) for x in [*np.linspace(0,25,10)]]], color="crimson", alpha=0.4, marker="o", linewidth=4.5, 
         markersize=10, label = "greedy")
plt.plot(bleu_beam, color="b", alpha=0.4, marker="o", linewidth=4.5, 
         markersize=10, label = "beam")
plt.xticks([*range(10)])

plt.ylabel("BLEU Score")
plt.xlabel("Epochs (*1/2)")
plt.rcParams["font.size"] = 16
plt.title("BLEU Score Progress by Epoch")

plt.legend()
plt.grid()
```
